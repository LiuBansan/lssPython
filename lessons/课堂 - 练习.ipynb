#### 体重，BMI 练习

```python
height = float(input("请输入你的身高（厘米）："))
weight = float(input("请输入你的体重（千克）："))

print("你的身高是：",height,"厘米")
print("你的体重是：",weight,"千克")

height /= 100
BMI = weight / (height ** 2)

if BMI < 18.5:
    print("你的BMI指数是：", int(BMI), "你的体重有点偏瘦")
elif BMI < 24.9:
    print("你的BMI指数是：", int(BMI), "你的体重属于正常")
elif BMI < 29.9:
    print("你的BMI指数是：", int(BMI), "你的体重有点偏重")
elif BMI > 29.9:
    print("你的BMI指数是：", int(BMI), "你的体重属于肥胖,该减肥啦")
```

####  判断某年是不是闰年，公历闰年遵循的规律为: 四年一闰,百年不闰,四百年再闰.

```python
year = int(input("请输入一个年份："))

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("{0}是闰年".format(year))      # 整百的年能被400整除是闰年
        else:
            print("{0}不是闰年".format(year))
    else:
        print("{0}是闰年".format(year))        # 非整百年能被4整除是闰年
else:
    print("{0}不是闰年".format(year))
```

#### 自然数求和 - 求1到N之间自然数的和

```python
N = int(input("请输入自然数N:"))

i = 0
sum = 0

while i <= N:
    sum += i
    i += 1
print("1到{0}之间自然数的和是{1}".format(N, sum))
```

#### 计算阶乘 - 计算1!+2!+3!+......+N!

```python
v = int(input("求v之前的阶乘之和,请输入自然数:"))

sum = 0
fac = 1

for i in range(1, v+1):
    fac = 1
    #strfac = strfac + "*" + str(i)
    
    for j in range(1,i+1):
        fac = fac * i
        
    sum += fac
    
print("{0}的阶乘是{1}".format(v,fac))
print("阶乘和",sum)
```

#### 鸡兔同笼  
- 今有鸡兔同笼，上有35头，下有94足，鸡兔分别有几只？

```python
chicken = 0
rabbit = 35 - chicken

while chicken <= 35:
    if 2 * chicken + 4 * rabbit == 94:
        print("鸡有{0}只，兔有{1}只".format(chicken,rabbit))
    chicken += 1
    rabbit = 35 - chicken
```

#### 百人分百饼
- 有一百个人，分一百个饼，其中，大人每人分3个，小孩子3人分一个，请问有多少个小孩，多少个大人。

```python
child = 0
adult = 100 - child

while child <= 100:
    if 3 * adult + 1/3 * child == 100:
        print("大人有{0}人，小孩有{1}人".format(adult,child))
    child += 1
    adult = 100 - child
```

```python
import random
mylist = []
for item in range(0, 10):
    mylist.append(random.randint(10, 100))
    
print("随机生成的列表为：",mylist)

maxnum = mylist[0]
minnum = mylist[0]

for item in mylist:
    if item < minnum:
        minnum = item
    if item > maxnum:
        maxnum = item
        
print("最大值是：",maxnum)
print("最小值是：",minnum)
print("-" * 15)
print("最大值是：",max(mylist))
print("最小值是：",min(mylist))
        
```

#### 冒泡排序
- 冒泡排序的思想: 每次比较两个相邻的元素, 如果他们的顺序错误就把他们交换位置。
- 冒泡排序原理: 每一趟只能将一个数归位, 如果有n个数进行排序,只需将n-1个数归位, 也就是说要进行n-1趟操作(已经归位的数不用再比较)
- 缺点: 冒泡排序解决了桶排序浪费空间的问题, 但是冒泡排序的效率特别低

```python
import random
mylist = []

# 给列表随机赋值
for item in range(0,10):
    mylist.append(random.randint(10,100))
print("随机生成的数值列表：{0}".format(mylist))

# 这个循环负责设置冒泡排序及逆行的次数
for i in range(len(mylist) - 1):
    
    # j 为列表下标
    for j in range(len(mylist) -i - 1):
        if mylist[j] > mylist[j+1]:
            mylist[j], mylist[j+1] = mylist[j+1],mylist[j]
    print("第", i+1 , "次排序后的数值列表：", mylist)
print("冒泡排序最后结果：",mylist)
    
```
